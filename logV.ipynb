```python
import json
import re
import pprint
```

```python
with open('dataset.json', 'r') as data:
    fullData = json.load(data)
```

```python
# dates
for date in fullData:
    print(date)
```

```python
# phases
phases = []
for date in fullData:
    for phase in fullData[date]:
        if phase not in phases:
            phases.append(phase)
            print(phase)
```

```python
# nodes
nodes = []
for date in fullData:
    for phase in fullData[date]:
        for node in fullData[date][phase]: 
            if node not in nodes:
                nodes.append(node)
                print(node)  
```

```python
# gpus - do not understand....are these diffrent GPUs or time stamps?
gpus = []
for date in fullData:
    for phase in fullData[date]:
        for node in fullData[date][phase]: 
            for gpu in fullData[date][phase][node]:
                if gpu not in gpus and gpu.startswith("GPU"):
                    gpus.append(gpu)
                    print(gpu)
```

```python
# "Aggregate" single and double bit flip totals 
sflips = 0
dflips = 0
for date in fullData:
    for phase in fullData[date]:
        for node in fullData[date][phase]: 
            for gpu in fullData[date][phase][node]:
                if gpu.startswith("GPU"):
                    sflips += int(fullData[date][phase][node][gpu]["ECC Errors"]["Aggregate"]["Single Bit"]["Total"])
                    dflips += int(fullData[date][phase][node][gpu]["ECC Errors"]["Aggregate"]["Double Bit"]["Total"])
print("Single Bit Flips:")                    
print(sflips)
print("Double Bit Flips:")                    
print(dflips)
```

```python
# "Aggregate" single and double bit flip totals by PHASE
phaseDict = {}
for date in fullData:
    for phase in fullData[date]:
        if phase not in phaseDict:
            phaseDict.update({phase: {'Single Flip Errors': 0, 'Double Flip Errors': 0}})
        for node in fullData[date][phase]: 
            for gpu in fullData[date][phase][node]:
                if gpu.startswith("GPU"):
                    phaseDict[phase]['Single Flip Errors'] += int(fullData[date][phase][node][gpu]["ECC Errors"]["Aggregate"]["Single Bit"]["Total"])
                    phaseDict[phase]['Double Flip Errors'] += int(fullData[date][phase][node][gpu]["ECC Errors"]["Aggregate"]["Double Bit"]["Total"])

for phase,values in phaseDict.items():
    print(phase)
    print(values)
    
```
